```python
data.reset_index(drop=True,inplace=True)
```

```python
X=data.loc[:,data.columns!='suicides_no'] #Setting dependent and independent variables
y=data.loc[:,data.columns=='suicides_no']
```

```python
from sklearn.preprocessing import StandardScaler #Scaling the data using Standard Scallar(mu=0 and standard_deviation=1)
Sc=StandardScaler()
```

```python
X=Sc.fit_transform(X) #Scaling
```

```python
from sklearn.model_selection import train_test_split #Splitting the data into 2 parts
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor 
lr=RandomForestRegressor(n_estimators=100,random_state=0)
```

```python
lr.fit(X_train,y_train) #Using the RandomForestRegressor to fit in the data.
```

```python
lr.score(X_test,y_test) 
```

```python
from sklearn.preprocessing import PolynomialFeatures
plf=PolynomialFeatures(degree=5)
```

```python
X_train=plf.fit_transform(X_train)
X_test=plf.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
poly=LinearRegression()
```

```python
poly.fit(X_train,y_train) #Using the Polynomial Regression to fit in the data.
```

```python
poly.score(X_test,y_test)
```

```python
from sklearn.ensemble import AdaBoostRegressor
adaboostModel=AdaBoostRegressor(n_estimators=50,base_estimator=RandomForestRegressor())
```

```python
adaboostModel.fit(X_train,y_train)
```

```python
adaboostModel.score(X_test,y_test)
```

```python
y_pred=adaboostModel.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred) #Getting the R-2 Score of out model.
```

```python
#Here we can see that RandomForest gave the accuracy score of 98.50% and adaboost gave of 99.39.

#Now choosing which model to use depends upon various cases like:

#1.Time:Random forest took a lot less time to fit as well as predict as it had only 100 trees.But Adaboost here has 50 
#estimators which each has 100 trees itself making them a lot time consuming in fitting as well as predicting than RandomForest.
#2.Memory:As you can see RandomForest has 100 tress and Adaboost has 50 of thoe 100 trees.RandomForest is memory efficient.
#3.Accuracy:If we really have the high accuracy requirement and we really don't care of time and space.Then we choose the
#Adaboost over RandomForest.

#First two cases we see more in real life scenarios and last very less times.
```
