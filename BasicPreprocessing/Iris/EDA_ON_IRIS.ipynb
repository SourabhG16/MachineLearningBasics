```python
#Importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Getting the data from the csv file using pandas library and storing into dataframe named df
df=pd.read_csv(r'C:\Users\Roggh\Iris.csv')
```

```python
#Shape is 150,6 so we have 6 columns(including ID) and 150 rows(Samples)
df.shape
```

```python
#Getting all columns
df.columns
```

```python
#Get the type of classes and their total count in the data 
df['Species'].value_counts()
#Data-set is balanced as all the classes are equally distributed across data(Generally they sould be newaly equally distributed) 
```

```python
#2-D Scatter Plot 
df.plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm')
plt.grid()
plt.show()
```

```python
#2-D color plot with the color coding
sns.set_style("whitegrid")
sns.FacetGrid(df,hue='Species',size=4).map(plt.scatter,"SepalLengthCm","SepalWidthCm").add_legend()
plt.show()
#As you can see by drawing the line we can easily seperate the class Iris-Setosa from the two other classes so it
#can be called as Linearly Sperable.The task of classification is easy when the classes are linearly seperable as
#it takes very less effort to find out the best sperating criteria for the classes as the criteria is as simple as Line.
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(df,hue='Species',size=4).map(plt.scatter,"PetalLengthCm","PetalWidthCm").add_legend()
plt.show()
#Again setosa is linearly sperable 
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(df,hue='Species',size=4).map(plt.scatter,"PetalLengthCm","SepalWidthCm").add_legend()
plt.show()
#Here setosa is seperable as well as we can see the versicolor and virginica can also be now classified to some extent.
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(df,hue='Species',size=4).map(plt.scatter,"SepalLengthCm","PetalWidthCm").add_legend()
plt.show()
#Same as above
```

```python
#Until now we have seen 2-D scatter plot which at a time can take 2 variables and plot them.Now we can plot more than 2 and
#see if we can draw a nicely seperable plane which seperates the data poits from each other according to their respective
#classes.If we plot 3-D data then we need a sperating plane same as we needed line to sperate 2-D data.
```

```python
df=df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm',
       'Species']]
```

```python
#PAIR-PLOTS
sns.set_style("whitegrid")
sns.pairplot(df,hue='Species',size=4)
plt.show()
#here we can see that we can now easily seperate the three flower types by using the variables and they are also
#linearly sperable
#combination of Petal length and petal width seems to be the best at seperating the classes
```
